```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[5,6,7],
              [1,3,2],
              [4,1,3],
              [8,9,7],
              [2,4,3],
              [7,7,6],
              [3,2,1],
              [6,8,8]])
y = np.array([1,0,0,1,0,1,0,1])
```

```python
def compute_cost ( x, y, w, b):
    
    m = x.shape[0]
    cost = 0
    for i in range(m):
        z = np.dot(w, x[i]) + b
        f_wb = 1 / (1 + np.exp(-z))
        
        cost += - y[i] * np.log(f_wb + 1e-8) - (1 - y[i]) * np.log(1 - f_wb + 1e-8)
        
    total_cost = cost / m
    return total_cost
```

```python
def compute_gradient ( x, y, w, b):

    m = x.shape[0]
    dj_dws = np.zeros_like(w)
    dj_dbs = 0
    for i in range(m):
        z = np.dot(w, x[i]) + b
        f_wb = 1 / ( 1 + np.exp(-z))

        dj_dw_i = (f_wb - y[i]) * x[i]
        dj_db_i = f_wb - y[i]

        dj_dws += dj_dw_i
        dj_dbs += dj_db_i
            
    dj_dw = dj_dws / m
    dj_db = dj_dbs / m

    return dj_dw, dj_db
```

```python
def gradient_descent ( x, y, w_in, b_in, iters, alpha, compute_cost, compute_gradient):
    w = w_in
    b = b_in
    cost_history = []
    for i in range(iters):
        dj_dw, dj_db = compute_gradient (x, y, w, b)

        w = w - alpha * dj_dw
        b = b - alpha * dj_db

        if i % 100 == 0:
            cost = compute_cost (x, y, w, b)
            cost_history.append(cost)
            print(f" w : {w} -- b : {b} = cost {cost}")
    return w, b, cost_history
            
```

```python
iters = 1000
alpha = 0.8
w_in = np.zeros(x.shape[1])
b_in = 0

final_w, final_b , cost_hist = gradient_descent(x, y, w_in, b_in, iters ,alpha, compute_cost, compute_gradient)

print(final_w, final_b)
```
